```python
import matplotlib.pyplot as plt
import numpy as np
from colorama import init
init()
from colorama import Fore, Back, Style
import pylab
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
r2=0.1
k=200*pow(10,-15)
h=10
vis=5/1000    
Pk=25
Rk=1500
q1=200/86400
q2=300/86400
rc=r2
```

```python
def Pc(q,k,h,Pk,vis,Rk,rc):
    ln=(np.log(Rk)-np.log(rc))/np.log(2.73)
    return Pk-((q*vis*ln)/(2*3.14*k*h))/pow(10,6)
```

```python
x = np.linspace(-1000, 1000, 100)
y = np.linspace(-1000, 1000, 100)
X, Y = np.meshgrid(x, y)
```

```python
P1=((Pk-Pc(q1,k,h,Pk,vis,Rk,rc))/((np.log(Rk)-
                                    np.log(rc))/
                                   np.log(2.73)))*((np.log(Rk)-
                                                    np.log(np.sqrt(abs((X+100)**2+
                                                                       (Y+100)**2))))/np.log(2.73))
P2=((Pk-Pc(q2,k,h,Pk,vis,Rk,rc))/((np.log(Rk)-
                                    np.log(rc))/
                                   np.log(2.73)))*((np.log(Rk)-
                                                    np.log(np.sqrt(abs((X-700)**2+
                                                                       (Y-500)**2))))/np.log(2.73))
P3=P1-P2
P=Pk-P3
```

```python
fig, axes = plt.subplots()

cs = plt.contour(X, Y, P,40)
cb = plt.colorbar()
cb.set_label('Pressure,')
ax.clabel(cs)
axes.set_xlabel('')
axes.set_ylabel('')
plt.show()
```
