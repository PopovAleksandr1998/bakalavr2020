```python
import matplotlib.pyplot as plt
import numpy as np
from colorama import init
init()
from colorama import Fore, Back, Style
import pylab
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
r2=0.1
k=200*pow(10,-15)
h=10
vis=5/1000    
Pk=25
Rk=1500
q=200/86400
rc=r2
```

```python
def Pc(q,k,h,Pk,vis,Rk,rc):
    ln=(np.log(Rk)-np.log(rc))/np.log(2.73)
    return Pk-((q*vis*ln)/(2*3.14*k*h))/pow(10,6)
```

```python
x = np.linspace(-1000, 1000, 100)
y = np.linspace(-1000, 1000, 100)
X, Y = np.meshgrid(x, y)
```

```python
P=Pk-((Pk-Pc(q,k,h,Pk,vis,Rk,rc))/((np.log(Rk)-
                                    np.log(rc))/
                                   np.log(2.73)))*((np.log(Rk)-
                                                    np.log(np.sqrt(abs((X+0)**2+
                                                                       (Y+0)**2))))/np.log(2.73))
```

```python
fig, axes = plt.subplots()

cs = plt.contour(X, Y, P,10)
cb = plt.colorbar()
cb.set_label('Pressure,МПа')
ax.clabel(cs)
ax.set_xlabel('Метры')
ax.set_ylabel('Метры')
plt.show()
```
